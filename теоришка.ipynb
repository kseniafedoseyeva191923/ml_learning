```python
porosity_logs = [0.12, 0.25, 0.08, 0.30, 0.15]
for p in porosity_logs:
    if p > 0.25:
        print("Отличный коллектор")
    elif p >= 0.15:
        print("Средний коллектор")
    else:
        print("Плотная порода")
```

```python
def check_reservoir (data_list):
    count = 0
    
    for p in data_list:
        if p > 0.2:
            count += 1
    return count

my_logs = [0.1, 0.3, 0.05, 0.4, 0.25]

result = check_reservoir(my_logs)

print("Количество хороших интервалов:", result)            
            
```
